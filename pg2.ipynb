```python
class Graph:
    def __init__(self,graph,heuList,start):
        self.graph=graph
        self.H=heuList
        self.start=start
        self.solGraph={}
        self.parent={}
        self.status={}
    
    def allpyaostar(self):
        self.aoStar(self.start,False)
    def getNeighbors(self,v):
        return self.graph.get(v,"")
    
    def getstatus(self,v):
        return self.status.get(v,0)
    def getheuristic(self,v):
        return self.H.get(v,0)
    def setstatus(self,v,val):
        self.status[v]=val
    def setheuristic(self,v,val):
        self.H[v]=val
        
    def printSol(self):
        print("Start from start node :" ,self.start)
        print("-------------------------------------------------------")
        print(self.solGraph)
        print("-------------------------------------------------------")
    
    def computeMin(self,v):
        mC=0
        mcL={}
        mcL[mC]=[]
        flag=True
        for nlt in self.getNeighbors(v):
            cost=0
            Nl=[]
            for c,weight in nlt:
                cost=cost+self.getheuristic(c)+weight
                Nl.append(c)
                
                if flag==True:
                    mC=cost
                    mcL[mC]=Nl
                    flag=False
                else:
                    if mC>cost:
                        mC=cost
                        mcL[mC]=Nl
        return mC,mcL[mC]
    
    def aoStar(self,v,backtracking):
        print("Heuristic values ",self.H)
        print("Soluctio ",self.solGraph)
        print("preocessing node ",v)
        print("---------------------------------------------------")
        
        if self.getstatus(v)>=0:
            minCost,childNL=self.computeMin(v)
            self.setheuristic(v,minCost)
            self.setstatus(v,len(childNL))
            solved=True
            
        for cN in childNL:
            self.parent[cN]=v
            if self.getstatus(cN)!=-1:
                solved=solved&False
        
        if solved==True:
            self.setstatus(v,-1)
            self.solGraph[v]=childNL
        
        if v!=self.start:
            self.aoStar(self.parent[v],True)
            
        if backtracking==False:
            for cN in childNL:
                self.setstatus(cN,0)
                self.aoStar(cN,False)
                
G1={
'A': [[('B', 1), ('C', 1)], [('D', 1)]],
'B': [[('G', 1)], [('H', 1)]],
'C': [[('J', 1)]],
'D': [[('E', 1), ('F', 1)]],
'G': [[('I', 1)]]
}
H1= {'A': 1, 'B': 6, 'C': 2, 'D': 12, 'E': 2, 'F': 1, 'G': 5, 'H': 7, 'I': 7, 'J': 1, 'T': 3}
g1=Graph(G1,H1,'A')
g1.allpyaostar()
g1.printSol()
                        
                    
                
            
        
    
    
```
