```python
import csv

def g_0(n):
    return ("?",)*n

def s_0(n):
    return ("$",)*n

def more_general(h1,h2):
    more_general_list=[]
    for x,y in zip(h1,h2):
        mg= x=="?" or (x!="$" and (x==y or y=="$"))
        more_general_list.append(mg)
    return all(more_general_list)

def fullfill(example ,hypothesis):
    more_general(hypothesis,example)
    
def min_generaliztion(h,x):
    h_new=list(h)
    for i in range(len(h)):
        if not fullfill(x[i:i+1],h[i:i+1]):
            h_new="?" if h[i]!="$" else x[i]
    return [tuple(h_new)]

def min_specialization(h,domain,x):
    result=[]
    for i in range(len(h)):
        if h[i]=="?":
            for val in domain[i]:

                    if val !=x[i]:
                        h_new=h[:i]+(val,)+h[i:]
                        result.append(h_new)
                    elif h[i]!="$":
                        h_new=h[:i]+("$",)+h[i:]
                        result.append(h_new)
    return result


with open('trainingexamplesT.csv') as csvFile:
    examples = [tuple(line) for line in csv.reader(csvFile)]
    


def get_domain(examples):
    d = [set() for i in examples[0]]
    for x in examples:
        for i,xi in enumerate(x):
            d[i].add(xi)
    return [list(sorted(x)) for x in d]


                       
def candidate_Elimination(examples):
    domains=get_domain(examples)
    G=set(g_0(len(domains)))
    S=set(s_0(len(domains)))
    i=0
    print("\nG[{0}]".format(i),G)
    print("\nS[{0}]".format(i),S)
    for xcx in examples:
        i=i+1
        x,xc=xcx[:-1],xcx[-1]
        if xc=="Y":
            G={g for g in G if fullfill(x,g)}
            S=generalize_S(x,G,S)
        else:
            S={s for s in S if not fullfill(x,s)}
            G=specialize_G(x,domains,G,S)
        
        print("\nG[{0}]".format(i),G)
        print("\nS[{0}]".format(i),S)
    return

def generalize_S(x,G,S):
    S_prev=list(S)
    for s  in S_prev:
        if s not in S:
            continue
        if not fullfill(x,s):
            S.remove(s)
            S_plus=min_generaliztion(s,x)
            S.update([h for h in S_plus if any([more_general(g,h) for g in G])])
            S.difference_update([h for h in S if any([more_general(h,h1) for h1 in S if h!=h1])])
    return S

def specialize_G(x,domain,G,S):
    G_prev=list(G)
    for g  in G_prev:
        if g not in G:
            continue
        if  fullfill(x,g):
            G.remove(s)
            G_minus=min_specialization(g,domain,x)
            G.update([h for h in G_minus if any([more_general(h,s) for s in S])])
            G.difference_update([h for h in G if any([more_general(g1,h) for h1 in G if h!=g1])])
    return G
                       
candidate_Elimination(examples)
                       
    
    
                       
```
